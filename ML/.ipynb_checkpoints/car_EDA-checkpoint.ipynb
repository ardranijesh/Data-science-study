```python
median_income = df['INCOME'].median()
median_travel_time = df['TRAVEL TIME'].median()
median_miles_clocked = df['MILES CLOCKED'].median()
mean_car_age = df['CAR AGE'].mean()
```

```python
df['INCOME'].replace(np.nan, median_income, inplace=True)
df['TRAVEL TIME'].replace(np.nan, median_travel_time, inplace=True)
df['MILES CLOCKED'].replace(np.nan, median_miles_clocked, inplace=True)
df['CAR AGE'].replace(np.nan, mean_car_age, inplace=True)
```

```python
object_columns= ['MARITAL STATUS', 'SEX', 'EDUCATION','JOB' ,'USE','CAR TYPE','CITY', 'POSTAL CODE']
```

```python
df['MARITAL STATUS'] = df['MARITAL STATUS'].fillna(df['MARITAL STATUS'].mode().values[0])
df['SEX'] = df['SEX'].fillna(df['SEX'].mode().values[0])
df['EDUCATION'] = df['EDUCATION'].fillna(df['EDUCATION'].mode().values[0])
df['JOB'] = df['JOB'].fillna(df['JOB'].mode().values[0])
df['USE'] = df['USE'].fillna(df['USE'].mode().values[0])
df['CAR TYPE'] = df['CAR TYPE'].fillna(df['CAR TYPE'].mode().values[0])
df['CITY'] = df['CITY'].fillna(df['CITY'].mode().values[0])
df['POSTAL CODE'] = df['POSTAL CODE'].fillna(df['POSTAL CODE'].mode().values[0])
```

```python
# Outlier
def remove_outlier(col):
    sorted(col)
    Q1, Q3 = col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    Lower_range = Q1 - 1.5 *IQR
    Upper_range =  Q1 + 1.5 *IQR
    return Lower_range , Upper_range
    
```

```python
lr_income, up_income = remove_outlier(df['INCOME'])
df['INCOME'] = np.where(df['INCOME']> up_income, up_income, df['INCOME'])
df['INCOME'] = np.where(df['INCOME']< lr_income, lr_income, df['INCOME'])
```

```python
lr_income, up_income = remove_outlier(df['INCOME'])
df['INCOME'] = np.where(df['INCOME']> up_income, up_income, df['INCOME'])
df['INCOME'] = np.where(df['INCOME']< lr_income, lr_income, df['INCOME'])
```

```python
lr_income, up_income = remove_outlier(df['TRAVEL TIME'])
df['TRAVEL TIME'] = np.where(df['TRAVEL TIME']> up_income, up_income, df['TRAVEL TIME'])
df['TRAVEL TIME'] = np.where(df['TRAVEL TIME']< lr_income, lr_income, df['TRAVEL TIME'])
```

```python
lr_income, up_income = remove_outlier(df['MILES CLOCKED'])
df['MILES CLOCKED'] = np.where(df['MILES CLOCKED']> up_income, up_income, df['MILES CLOCKED'])
df['MILES CLOCKED'] = np.where(df['MILES CLOCKED']< lr_income, lr_income, df['MILES CLOCKED'])
```

```python
lr_income, up_income = remove_outlier(df['CAR AGE'])
df['CAR AGE'] = np.where(df['CAR AGE']> up_income, up_income, df['CAR AGE'])
df['CAR AGE'] = np.where(df['CAR AGE']< lr_income, lr_income, df['CAR AGE'])
```
